```python
from glob import glob
import time
```

```python
all_path = glob(filepath+'*.csv') + glob(filepath+'*.xlsx')
len(all_path)
```

```python
# 파일 읽기 시작 시간 기록
start_time = time.time()

df = pd.read_csv(all_path[0], encoding='cp949')

# 파일 읽기 종료 시간 기록
end_time = time.time()

# 걸린 시간 계산
time_taken = end_time - start_time
print(len(df))
print(time_taken)
print(f'하나의 row를 읽는 시간은:{time_taken / len(df)}')
```

```python
csv_rows = 6894866
excel_rows = 465715

# 초당 처리된 행수
print(csv_rows // (t2-t1))
print(excel_rows // (t3-t2))
```

```python
all_cols = []

for path in all_path:
    if '.csv' in path:
       cols = list(pd.read_csv(path, encoding='cp949', skiprows=lambda x: x>2).columns) # 3이상의 로우(행)는 생략해줘
    else:
       cols = list(pd.read_excel(path, skiprows=lambda x: x>2).columns)

    all_cols = all_cols + cols
    
set(all_cols)
```

```python
reset_col = all_cols[:12]
reset_col
```

```python
reset_col = all_cols[:12]
interested = ['대여일자', '대여시간', '대여구분코드', '성별', '연령대코드']

all_df = pd.DataFrame()
for path in all_path:
    if '.csv' in path:
        single = pd.read_csv(path, encoding='cp949')
    else:
        single = pd.read_excel(path)

    single.columns = reset_col
    all_df = pd.concat([all_df, single[interested]])
    print(all_df.shape)
    
    del single # 기존에 있던 single을 삭제해야 이어붙일때 기존 데이터를 안가져옴
```

```python
all_df.to_csv('test1_modify.csv', index=False)
```

```python
# regular, casual, group

mapper = {'단기권(전기)': 'casual',
          '단체': 'group',
          '단체권': 'group',
          '일일(2시간권)': 'casual',
          '일일(비회원)': 'casual',
          '일일(회원)': 'casual',
          '일일권': 'casual',
          '일일권(비회원)': 'casual',
          '정기': 'regular',
          '정기(2시간권)': 'regular',
          '정기권': 'regular'}

df['type'] = df['대여구분코드'].map(mapper)
```

```python
# F, M, U(모르겠어)

mapper = {'': 'U', 
          'F': 'F', 
          'M': 'M', 
          '\\N': 'U', 
          'f': 'F', 
          'm': 'M', 
          'nan': 'U'}

df['gender'] = df['성별'].map(mapper)
set(df['gender'])
```
