## 데이터 분석 방법론
- 전통적인 데이터 분석 방법론 : KDD, CRISP-DM
- 체계화된 절차와 방법이 정리된 분석 방법론 필요
- 방법론은 절차와 방법, 도구, 기법, 템플릿과 산출물로 구성되어 있음

### CRISP-DM
- Cross Industry Standard Process - Data Mining
- 가장 많이 사용되는 Data Mining 표준 방법론
- 분야, 산업군을 막론하고 데이터로부터 의미를 도출하는 표준 절차

#### Business Understanding
- 분석할 분야의 목적, 요구사항을 이해하고, 초기의 프로젝트 계획을 수립하는 단계
- 업무 목적과 상황 파악, Data Mining의 목표 설정, 프로젝트 계획 수립 등의 세부 업무 포함
- 해당 분야의 종사자들이 원하는 방향
- 분야 별로 관심사(KPI)가 다름
- 제조업(공장): 매출액 증대, 재구매율 증가
- 숙박업(호텔): 매출 극대화, 운영의
- 과거 학술, 기술적인 선행연구가 무었이었는지를 파악
- **분석 방향성** 및 목표 설정(업계의 고민 해결에 도움이 되는 목표나 방향)


#### Data Understanding
- 데이터를 수집하고, 데이터 속성을 이해하기 위한 단계
- 데이터 기술 분석, EDA, 데이터의 품질을 확인함
- 가용 데이터로 기존에 설정한 분석 방향성 및 목표 달성이 가능할지 파악하는 단계


EDA : Exploratory Data Analysis, 탐색적 데이터 분석
- 시각화를 통해 데이터 안의 유효한 패턴을 찾음
- 수집한 데이터를 관찰하고 이해하기 위한 단계
- 데이터를 분석하기 전에 그래프나 통계적인 방법으로 자료를 직관적으로 바라보는 단계
- Business Understanding에서 설정한 방향성과 목표에 부합하면 유효

1. 데이터의 출처와 주제에 대한 이해
2. 데이터 크기 확인
3. 데이터의 구성요소 확인(feature의 속성 탐색)
4. 탐색한 데이터의 시각화

##### 데이터 품질
- 결축치, 일관성, 형태, 정보

#### Data Preperation
- 분석 기법에 적합한 데이터를 편성하는 단계
- 데이터 셋 선택, 데이터 정제 등
- 모델링(Machine Learning)을 대비하는 단계
- 알고리즘에 넣을 데이터를 '적절하게' 준비
- 데이터를 알고리즘에 넣기 위해 형태를 맞추는 단계

##### Machine Learning
- 데이터의 패턴을 알고리즘이 파악
- 학습에 방식에 따라 지도학습과 비지도학습으로 나뉨
- 지도학습 : 문제(feature)와 답(lable)을 줌
- 비지도학습 : 문제만 줌
- 학습용(train)과 시험용(test) 데이터셋 분리(split)
- 학습했던 문제가 시험문제에 똑같이 나오면 시험 점수를 신뢰할 수 없음
- 데이터 셋의 분리는 공부했던 문제가 시험문제에 나오지 않게(중복되지 않게)하기 위함
- feature - lable split 

#### Modeling
- 다양한 모델링 기법과 알고리즘을 선택하고, 사용되는 파라미터들을 최적화 하는 단계
- 모델의 성능을 평가하고, Overfitting(과적합) 되었는지 확인
- Model Planning, Model Building

##### Model Planning
- 알고리즘 후보군 중에서 선정
- 속도, 예측력, 해석력을 고려해서 알고리즘 후보군을 선정
- 내부 계산과정이 복잡할수록 속도↓, 예측력↑, 해석력↓
- 일반적으로 기술통계 분야 알고리즘은 속도와 해석력 우수함

##### Model Building
- 모형 수립, 학습, 평가지표 적용
- Model Planning에서 고른 알고리즘에 학습용 데이터로 학습하고 시험용 데이터로 채점
- 채점 시 관점에 따라서 평가지표가 달라짐
- case1 ) 정답 1 예측값 2
- case2 ) 정답 1001 예측값 1002

##### 평가지표
- 서로 다른 알고리즘 중에서 한 평가지표 기준으로 어떤 알고리즘의 정확도가 좋은가
- 한 알고리즘에 대해 옵션을 설정하여 조금씩 다른 알고리즘을 만들었을 때 상호비교용

#### Evalution
- Modeling의 결과가 프로젝트 목적에 부합하는지를 평가
- Modeling 가정, 분석 결과 등
- Business Understanding에서 설정한 방향성이나 목표를 달성했는지 파악
- Executive Summary, Technical Report 작성

#### Deployment
- 완성된 모델을 실제 업무에 적용하기 위한 단계
- Operationalize
- 모니터링과 모델의 유지보수 계획을 마련
