### 그래프에서 어떤 속성을 줘야할까

```python
plt.plot([1,2,3],[3,6,9], marker='o', linestyle='-.', color='g')
plt.plot([1,2,3],[1,2,3], marker='^', linestyle=':', color='m')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.title('제목')
plt.xlabel('x 레이블')
plt.ylabel('y 레이블')
plt.xticks(ticks=[1,2,3], rotation=20)
plt.yticks(ticks=[i for i in range(0, 11, 2)])
plt.legend(['legend1', 'legend2'])

plt.grid()
plt.show()
```

#### 사분위수
- 자료를 크기 순으로 벼열하고, 누적 백분율을 4등분한 각 점에 해당하는 값
- 제 1 사분위 수 : 누적 백분율이 25%
- 제 2 사분위 수 : 누적 백분율이 50%
- 제 3 사분위 수 : 누적 백분율이 75%
- 제 4 사분위 수 : 누적 백분율이 100%
- 박스의 중간에 있는 선이 2사분위 수
- 박스의 아래 테두리가 1사분위 수
- 박스의 위 테두리가 3사분위 수
- 막대기는 최대 최소 값
- 가장 위 점은 이상치(outlier) : 통계에는 적절하지 않는 수치다.
- 이상치 처리법: 이상치를 없엔다, 최대 최소값으로 전환한다,
