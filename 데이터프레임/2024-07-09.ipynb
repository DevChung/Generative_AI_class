```python
df['weather'].value_counts() # 각 값의 나온 횟수를 세줌 AI학습의 셈플 갯수를 볼때 사용해야 함 중요!!!
```

```python
# 상관계수(coefficient of correlation)
# 두 변위를 측정했을 때 한 변인의 변화에 따라 그의 대응하는 다른 변인이 어떻게 변하는가?
df[['season', 'temp']].corr()  # 0.4이하 의미없다
# 0.5 중간
# 0.66 이상 상관관계가 높은 편이다
```
