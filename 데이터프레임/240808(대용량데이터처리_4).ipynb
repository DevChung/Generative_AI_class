```python
# humidex(습도지수)
# DI(습도지수)=0.81×T+0.01×RH×(0.99×T−14.3)+46.3
# T: 섭씨온도, RH: 상대습도

# 20 이하: 쾌적함을 느끼는 상태 -> 0
# 21-29: 약간 더운 상태로, 대부분의 사람들이 불편함을 느끼지 않음 -> 1
# 30-39: 더운 상태로, 불편함을 느끼기 시작할 수 있음 -> 2
# 40-45: 매우 더운 상태로, 상당한 불쾌감을 느낌 -> 3
# 45 이상: 위험한 수준으로, 열사병 등의 건강 위험이 증가할 수 있음 -> 4
```

```python
df['DI'] = 0.81 * df['temp'] + 0.01 * df['humid'] * (0.99 * df['temp'] - 14.3) + 46.3
```

```python
df['DI'] = df['DI'].round(2)
```

```python
set(df['DI'])
```

```python
# 습도지수 기준에 따라 구간 설정 및 라벨 지정
bins = [-float('inf'), 20, 29, 39, 44, float('inf')]
labels = [0, 1, 2, 3, 4]

# pd.cut()을 사용하여 DI 열 변경
df['DI_category'] = pd.cut(df['DI'], bins=bins, labels=labels)

# (-inf, 20]: 20 이하
# (20, 29]: 21부터 29까지
# (29, 39]: 30부터 39까지
# (39, 44]: 40부터 44까지
# (44, inf): 45 이상
# 왼쪽 열고 오른쪽은 닫힘, right=False옵션 주면, 오른쪽도 열린다
print(df)
```

```python
# Heat Index(열지수)
```

```python
def calculate_heat_index(row):
    # 화씨로 변환된 온도
    temp_f = row['temp'] * (9/5) + 32
    humid = row['humid']
    
    # 기온이 80°F 이상일 때만 열지수 계산
    if temp_f >= 80:
        T2 = temp_f * temp_f
        RH2 = humid * humid
        T_H = temp_f * humid
        T2_H = T2 * humid
        T_H2 = temp_f * RH2
        T2_H2 = T2 * RH2

        heat_index = (-42.379
                      + 2.04901523 * temp_f
                      + 10.14333127 * humid
                      - 0.22475541 * T_H
                      - 0.00683783 * T2
                      - 0.05481717 * RH2
                      + 0.00122874 * T2_H
                      + 0.00085282 * T_H2
                      - 0.00000199 * T2_H2)
        return heat_index
    else:
        return 0  # 기온이 80°F 미만일 경우 열지수 계산하지 않음

# 데이터프레임에 열지수 계산 적용
df['heat_index'] = df.apply(calculate_heat_index, axis=1)
```

#### 날씨가 사람의 건강에 미치는 영향
- 여름엔 열사병(heat index, 열 지수), 겨울엔 동상 위험(windchill, 겨울 체감온도)
- 열지수가 높으면 열사병 걸릴 확률이 증가 -> 외부활동 저하 요인
- 겨울 체감온도는 동상 확률과 연관 -> 외부활동 저하 요인
- 열지수나 겨울 체감온도는 도메인의 특성상 자전거를 이용한 외부활동이므로 고려해야 함

```python
# heat index
```

```python
def calculate_heat_index(temp_celsius, relative_humidity):
    # 섭씨를 화씨로 변환
    temp_fahrenheit = temp_celsius * 9/5 + 32

    # 열지수 계산을 위한 상수
    c1 = -42.379
    c2 = 2.04901523
    c3 = 10.14333127
    c4 = -0.22475541
    c5 = -6.83783e-3
    c6 = -5.481717e-2
    c7 = 1.22874e-3
    c8 = 8.5282e-4
    c9 = -1.99e-6

    # 열지수 계산
    heat_index_fahrenheit = (c1 + (c2 * temp_fahrenheit) + (c3 * relative_humidity) +
                             (c4 * temp_fahrenheit * relative_humidity) +
                             (c5 * temp_fahrenheit**2) +
                             (c6 * relative_humidity**2) +
                             (c7 * temp_fahrenheit**2 * relative_humidity) +
                             (c8 * temp_fahrenheit * relative_humidity**2) +
                             (c9 * temp_fahrenheit**2 * relative_humidity**2))

    # 화씨를 섭씨로 변환
    heat_index_celsius = (heat_index_fahrenheit - 32) * 5/9

    return heat_index_celsius

# 데이터프레임에 열지수 계산 적용
rcg['heat_index'] = rcg.apply(lambda row: calculate_heat_index(row['temp'], row['humid']), axis=1)

# 
```
