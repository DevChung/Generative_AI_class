```python
df_combined.pivot_table(index='User-ID', columns='Book-Title', values='Book-Rating')
## 데이터가 너무 커서 피벗 테이블을 생성할 수 없음.(indexError)
# 데이터 : 고객 수 * 책의 권수 = 9.2만 * 24만 => column수에 영향이 있음
# 어떻게 해결하면 좋을까?
# 연결 고리가 없는 것은 의사결정에 기여하기 어려움 -> 연결고리가 없다는 것 -> 라이트 유저, 인기없는 도서
# 연결고리가 있는 것-> 헤비 유저, 베스트 셀러 -> 극소수
# 의사결정 과정에서 기여를 못하는 Item과 유저를 적절히 필터링 해줘야한다.
```

```python
df_combined['count'] = 1
user_c = df_combined.groupby(by='User-ID').agg({'count':'sum'}).sort_values(by='count', ascending=False).reset_index()
user_c
```

```python
user_c['count'].hist(bins=100, log=True) 
# 너무 많이 1권에 모여 있어서 log스케일로 표현
# 적어도 많아도 다 표시 가능
```

```python
len(user_c.loc[user_c['count'] <= 100])
heavy_user = user_c.loc[user_c['count'] > 100]['User-ID']
```

```python
# best seller
book_c = df_combined.groupby(by='Book-Title').agg({'count':'sum'}).sort_values(by='count', ascending=False).reset_index()
book_c
```

```python
book_c.hist(bins=100, log=True)
# pareto distribution: 부의 분포를 나타내기 위한 용도로 사용
# 사회에서는 부의 불공평한 부포로 인해 대부분 부가 소수에 의해 쇼유된다는 그래프

#  youtube 에서 조회수가 백만인 컨텐츠는 무조건 댓글의 수가 백만 개는 아님
# 대부분 댓글을 안 달거나 하나 정도 달 것이고,
# 댓글을 많이 달면 달수록 그 인원 수는 점점 극소수가 딜 것임
```

```python
len(book_c.loc[book_c['count'] <= 10])
pop_book = (book_c.loc[book_c['count'] > 10]['Book-Title'])
```
