```python
all_cols = []
for path in all_path:
    if '.csv' in path:
        cols = list(pd.read_csv(path, encoding='cp949', skiprows=lambda x: x>2))
    elif '.xlsx' in path:
        cols = list(pd.read_excel(path, skiprows=lambda x: x>2))
    
    all_cols += cols
```

```python
set(all_cols)
```

```python
reset_col = all_cols[:12]
interest = ['대여일자', '대여시간', '대여구분코드', '성별', '연령대코드']

all_df = pd.DataFrame()
start = time.time()
for path in all_path:
    if '.csv' in path:
        single = pd.read_csv(path, encoding='cp949')
    else :
        single = pd.read_excel(path)

    single.columns = reset_col
    all_df = pd.concat([all_df, single[interest]])

    del single

end = time.time()
```

```python
for col in df.columns:
    df[col] = df[col].apply(lambda x: str(x).replace("'",''))
```

```python
mapper = {'단기권(전기)': 'casual', '단체':'group',  '단체권':'group', 
          '일일(2시간권)':'casual', '일일(비회원)':'casual', '일일(회원)':'casual',
 '일일권':'casual', '일일권(비회원)':'casual', '정기':'regular', '정기(2시간권)':'regular', '정기권':'regular'}
df['type'] = df['대여구분코드'].map(mapper)
```

```python
set(df['type'])
```

```python
# F : 여성, M : 남성, U : 불명
mapper = {'':'U',
          'F':'F',
          'M':'M',
          '\\N':'U',
          'f':'F',
          'm':'M',
          'nan':'U'}
df['gender'] = df['성별'].map(mapper)
```

```python
set(df['gender'])
```

```python
set(df['연령대코드'])
```

```python
mapper = {'20대':'20s',
          '30대':'30s',
          '40대':'40s',
          '50대':'50s',
          '60대':'60s',
          '70대~':'70s',
          'nan':'unknown', 
          '~10대':'10s'}
df['age'] = df['연령대코드'].map(mapper)
```
