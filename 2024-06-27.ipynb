### WEB Crawling에 대한 정확한 이해
- Web + Crawl + ing

#### 느리다는 단점에도 불구하고 selenium을 사용하는 이유
- selenium은 이미지를 불러와야 함 -> requests보다 느림(출력안함)
- selenium은 타이핑과 클릭이 가능함
- ex) 웹 사이트에서 특정 검색어를 검색한 후, 그에 대한 결과를 얻고자 하는 경우

#### 타이핑 절차
1. 원하는 지점을 찾는다
2. 원하는 지점을 선택한다
3. 원하는 정보를 입력한다

#### 클릭 절차
1. 원하는 원소를 찾는다
2. 원하는 원소를 선택한다
3. 클릭() 을 한다.
 

### 클릭과 타이핑이 공통으로 적용되는 일반화 된 문장
1. 화면을 구성하는 구성요소(element)를 찾음 -> A
2. 구성요소에 대해서 행위를 함 -> B
3. a = ? 로 A를 선택하고, a.B() 를 사용하여 A에게 B라는 메서드를 실행함

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
url = "https://www.aladin.co.kr/home/wusedshopmain.aspx?start=we"

driver = webdriver.Chrome()
time.sleep(1)
driver.get(url)
time.sleep(1)
search_box = driver.find_element(By.XPATH , '//*[@id="SearchWord"]')

search_box.send_keys('해리포터')
# search_box.send_keys(Keys.ENTER)

search_button = driver.find_element(By.XPATH, '//*[@id="global_search"]/button')
search_button.click()
time.sleep(1)

response = driver.page_source  # requests 에서 get().text 같은 기능
driver.close()

# <input type="text" id="SearchWord" name="SearchWord" class="inputbox_new" 
# style="border:0px solid white; ime-mode:active;" value="" onclick="javascript: searchChk();">
```

#### 베스트셀러 값찾기
1. 베스트 셀러 책 제목을 수집
2. 각각의 베셀 책 제목을 온라인 중고 페이지에 검색
3. 중고 서적의 정보를 수집

```python
def find_bestseller_top50(page=1):
    
    driver = webdriver.Chrome()
    time.sleep(1)
    
    for i in range(1, page+1):
        url = f'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page={i}&cnt=1000&SortOrder=1'
        driver.get(url)
        time.sleep(1)
        response = driver.page_source  # requests 에서 get().text 같은 기능
    
        one_page = bs4.BeautifulSoup(response)
        title_list = one_page.find_all('b')
        
        for title in title_list[0::3]:
            # print(title.text)
            find_wusedshop(title.text, driver)

    driver.close()
```

```python
def find_wusedshop(book_title, driver):
    url = "https://www.aladin.co.kr/home/wusedshopmain.aspx?start=we"
    driver.get(url)
    time.sleep(1)
    search_box = driver.find_element(By.XPATH , '//*[@id="SearchWord"]')
    
    search_box.send_keys(book_title)
    # search_box.send_keys(Keys.ENTER)
    
    search_button = driver.find_element(By.XPATH, '//*[@id="global_search"]/button')
    search_button.click()
    time.sleep(1)
    
    response = driver.page_source  # requests 에서 get().text 같은 기능

    buetify_book_info(response, book_title)
```

```python
import pandas as pd

def buetify_book_info(response, title):
    book_page = bs4.BeautifulSoup(response)
    try:
        book_area = book_page.find('div', id='Search3_Result')
        book_list = book_area.find_all('div', 'ss_book_box')
        book_dataframe = []
        
        for book in book_list:
            book_total_data = {}
            
            book_total_data['book_title'] = book.find('a', 'bo3').text
            # book_total_data['new_price'] = book.find_all('tr')[1].find_all('a', 'bo_used')[0].text
            # book_total_data['now_price'] = book.find_all('tr')[1].find_all('a', 'bo_used')[1].text
    
            book_dataframe.append(book_total_data)
    
    
        df = pd.DataFrame(book_dataframe)
        df.to_csv(title+'.csv', encoding='utf-8-sig', index=False) 
    
        print(f"{title}.csv가 생성되었습니다")

    except:

        print(f"{title}의 중고책장은 비어있습니다")
        return None
   
    
    
    
```
