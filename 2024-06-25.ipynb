### 2진수 
왼쪽 첫번째 비트는 부호비트
0000 0101   +5

- 1111 1010   1의 보수
- 1111 1011   2의 보수 ==> -5

2의 보수를 해야 음수로 취급됨

- 0000 0101
- 1111 1011   ==> 더하면 0이됨

과제 다음 이진수의 10진수를 구하시오

- 1000 0000 -128       0111 1111  1000 0000 128
- 1010 1100 -84        0101 0011  0101 0100 84



# 뉴스 정보를 수집/ 저장하기

```python
import requests
import xmltodict
import pandas as pd
```

```python
def news_get(keyword):
    url=f"https://news.google.com/rss/search?q={keyword}&hl=ko&gl=KR&ceid=KR%3Ako"
    response = requests.get(url).text
    news_data = xmltodict.parse(response)
    news_list = news_data['rss']['channel']['item']
    df = pd.DataFrame(news_list)
    df.to_csv(keyword+'.csv', index=False, encoding='utf-8-sig')
```

```python
news_get("반도체")
```

```python
import bs4

url = "https://www.weather.go.kr/w/rss/dfs/hr1-forecast.do?zone=4413125000"
response = requests.get(url).text

data = bs4.BeautifulSoup(response, features="xml")
data.find('wfKor')
```

```python
weather_list = data.find('body').find_all('data')
```

```python
weather_list
```

```python
import pandas as pd

weather_data=[]

for weather in weather_list:
    day = weather.find('day').text
    hour = weather.find('hour').text
    wfKor = weather.find('wfKor').text

    weather_dict = {'day':day, 'hour' : hour, 'wfKor':wfKor}

    weather_data.append(weather_dict)
```

```python
    pd.DataFrame(weather_data).to_csv('weather.csv', index=False, encoding='utf-8-sig')
```

### 기본적인 Web Crawling 절차
1. 요청할 요청 준비
2. get 요청을 보내고, 응답을 받아옴
3. 응답의 종류(Json, xml, html)에 따라서 해석의 도구가 달라짐
4. 해석방식(Dictionary 식으로 해석, tag 개념으로 해석)에 따라서 프로세스가 달라짐
5. 프로세스에 맞게 데이터를 정리 정돈 -> DataFrame화
6. 파일(.csv)로 저장

### xml과 html의 차이점
- xml은 태그의 이름이 태그 안에 있는 정보의 제목(대표성 있는 단어)역할을 한다
- HTML은 태그 이름이 태그 안에 있는 정보의 제목역할을 하지 않음


- HTML은 태그 이름이 화면의 디자인 또는 기능의 역할을 할 수 있다.
- ex. strong 태그는 글자를 진하게 함, h1....h5 태그 글자의 크기를 나타냄

```python
import pandas as pd

def search_news(keyword):
    news_output = []
    
    url = f"https://news.nate.com/search?q={keyword}"
    response = requests.get(url).text
    news_data = bs4.BeautifulSoup(response)
    news_list = news_data.find('ul', 'search-list').find_all('li', 'items')

    # print(news_list[0])
    # print("--------------------------------------------------------------------")
    # print(news_list[1])

    for news in news_list:
        news_title = news.find('h2').text
        news_text = news.find('span', 'txt').text
        news_link = "https:" + news.find('a').attrs['href']
        
        news_dict = {'title': news_title, 'text':news_text, 'link':news_link}
        news_output.append(news_dict)

    pd.DataFrame(news_output).to_csv(keyword+".csv", index=False, encoding='utf-8-sig')
    
    
    # news_dict = 
    # print(news_list[0])    
```

```python
import time

def get_top10_news():
    url = "https://news.nate.com/search?q=q"    
    response = requests.get(url).text
    realtime_word_data = bs4.BeautifulSoup(response)
    realtime_word_list = realtime_word_data.find('span', 'kwd-list').find_all('a')

    for title in realtime_word_list:
        search_news(title.text)
        print(title.text+".csv가 생성되었습니다")
        time.sleep(1)

    
```
